```python
##############################
        # Labels
############################
```

```python
x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,320,330])
plt.plot(x, y)
plt.title("Linear")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```

```python
##############################
        # Subplots
############################
```

```python
# plot 1
x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,320,330])
plt.subplot(1,2,1)
plt.title("1")
plt.plot(x,y)
# plot 2
x = np.array([1,23,4,56,7,8,9,15,153,165])
y = np.array([240,250,260,270,280,290,300,310,320,330])
plt.subplot(1,2,2)
plt.title("2")
plt.plot(x,y)
```
