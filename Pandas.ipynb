```python
# Aggregation & Grouping (Toplulaştırma ve Gruplama)
```

```python
df.groupby("sex").agg({"age" :["sum","mean"]})
```

```python
df.groupby("sex").agg({"age" :["sum","mean"],"survived":"mean"})
```

```python
df.groupby(["sex","embark_town"]).agg({"age" :["sum","mean"],"survived":"mean"})
```

```python
df.groupby(["sex","embark_town","class"]).agg({"age" :["sum","mean"],"survived":"mean"})
```

```python
df.groupby(["sex","embark_town","class"]).agg({"age" :["sum","mean"],
                                               "survived":"mean",
                                              "sex":"count"})
```

```python
df1 = pd.DataFrame({"employees":["john","dennis","mark","maria"],
                   "group" : ["accounting","engineering","engineering","hr"]})
df2 =  pd.DataFrame({"employees":["mark","dennis","john","maria"],
                   "start_date" : [2010,2009,2014,2019]})
```

```python
pd.merge(df1,df2,on = "employees")
```

```python
df3 = pd.merge(df1,df2)
```

```python
df4 = pd.DataFrame({"group": ["accounting","engineering","he"],
                   "manager": ["Caner","Mustafa","Berkcan"]})
```

```python
pd.merge(df3,df4)
```
